```python
def most_genre(list_in):
    genre_list = {'happy':0, 'sad':0, 'angry':0, 'calm':0}
    
    for i in list_in:
        if i == 'happy':
            genre_list['happy'] += 1
        if i == 'sad':
            genre_list['sad'] += 1
        if i == 'angry':
            genre_list['angry'] += 1
        if i == 'calm':
            genre_list['calm'] += 1
            
    return(max(genre_list, key=genre_list.get))
```
